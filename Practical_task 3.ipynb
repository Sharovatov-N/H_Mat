1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def pol_cart(rho, phi):
    x = rho * np.cos(math.radians(phi))
    y = rho * np.sin(math.radians(phi))
    return x, y
pol_cart(13, 45)
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 =  fsolve(equations, (-2, 2))
              
print (x1, y1)

x2, y2 =  fsolve(equations, (2, 5))

print (x2, y2)
```
